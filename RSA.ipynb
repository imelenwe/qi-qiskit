```python
from math import gcd
from sympy import randprime

'''
This code implements a simple RSA encryption and decryption scheme. It generates a pair of RSA keys, encrypts a message using the public key, and decrypts it using the private key.

1. Choose p and q randomly from the list of prime numbers. These will be the two prime factors of n.
2. Compute n = p * q. This will be the modulus for both the public and private keys.
3. Compute phi = (p - 1) * (q - 1). This is the totient of n.
4. Choose an integer e such that 1 < e < phi and gcd(e, phi) = 1. This will be the public exponent. A common choice for e is 3, but it can be any odd integer less than phi that is coprime to phi.
5. Compute d such that d ≡ e^(-1) (mod phi). This will be the private exponent. It can be computed using the Extended Euclidean Algorithm to find the multiplicative inverse of e modulo phi. The public key is (n, e) and the private key is (n, d).
'''

def gen_rsa_keypair():
    p = randprime(13, 100) # we use ord(c) < 128, so p and q should be > 128 works for both smallcase and uppercase letters
    q = randprime(13, 100)
    print(f"p={p}, q={q}")
    n = p * q
    phi = (p - 1) * (q - 1)

    e = 3
    while gcd(e, phi) != 1:
        e += 2

    d = pow(e, -1, phi) # mod operation to find multiplicative inverse of e mod phi
    return (n, e), (n, d)


def encrypt(message, pub_key):
    n, e = pub_key
    return [pow(ord(c), e, n) for c in message] # m^e mod n

def decrypt(ciphertext, priv_key):
    n, d = priv_key
    return ''.join([chr(pow(c, d, n)) for c in ciphertext]) # c^d mod n
```

```python
# test
pub, priv = gen_rsa_keypair()
n, e = pub
_, d = priv
print("testing keypair...")
print(f"n={n}, e={e}, d={d}")
print(f"pub key: {pub}")
print(f"priv key: {priv}")

encrypted = encrypt("Let us break RSA", pub)
print(f"encrypted: {encrypted}")

decrypted = decrypt(encrypted, priv)
print(f"decrypted: {decrypted}")
```
