```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector, plot_histogram
%matplotlib inline
import math
from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(2,1)
qc.h(0)
qc.x(1)
qc.h(1)
qc.barrier()
qc.draw(output="mpl")
```

```python
qc.cx(0,1)
qc.barrier()
qc.h(0)
qc.barrier()
qc.draw(output="mpl")
```

```python
qc.measure(0,0)
qc.barrier()
qc.draw(output="mpl")
```

```python
qasm_sim = AerSimulator()
result = qasm_sim.run(qc, shots=1).result()
counts = result.get_counts(qc)
print(counts)
#The counts variable is a Python dictionary that stores the results of your quantum experiment.
#Keys: The keys of the dictionary are strings, where each string represents a classical bit 
#outcome that was measured. For example, '010' or '1000101'.
#Values: The values are integers, telling you how many times each specific outcome occurred 
#over all the "shots."
```
